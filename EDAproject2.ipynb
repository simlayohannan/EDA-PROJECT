### 2. What is the age distribution of account holders, and how does it vary by account type?

```python
df=pd.read_csv(r"C:\Users\LENOVO\OneDrive\Data_analysis\dataset\bank_account_dataset.csv")
from datetime import datetime

# Calculate the age of account holders
current_year = datetime.now().year
df['Age'] = current_year - pd.to_datetime(df['Date_of_Birth']).dt.year

# Group the data by Account_Type and calculate age statistics
age_distribution_by_account_type = df.groupby('Account_Type')['Age'].describe()

# Display the age distribution
print(age_distribution_by_account_type)
```

### 3.What is the average balance across different account types and currencies?


```python

df=pd.read_csv(r"C:\Users\LENOVO\OneDrive\Data_analysis\dataset\bank_account_dataset.csv")
# Calculate the average balance grouped by Account_Type and Currency
average_balance = df.groupby(['Account_Type', 'Currency'])['Balance'].mean().reset_index()

# Visualization
plt.figure(figsize=(12, 6))
sns.barplot(
    data=average_balance,
    x='Account_Type',
    y='Balance',
    hue='Currency',
    palette='viridis'
)
plt.title('Average Balance by Account Type and Currency', fontsize=16)
plt.xlabel('Account Type', fontsize=12)
plt.ylabel('Average Balance', fontsize=12)
plt.legend(title='Currency')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

### 4. How does the credit score distribution vary across account statuses (Active, Dormant, Closed)?

### 5. Which branches have the highest average balances or transaction counts?

```python
df=pd.read_csv(r"C:\Users\LENOVO\OneDrive\Data_analysis\dataset\bank_account_dataset.csv")
# Calculate the highest average balances by branch
highest_avg_balance = df.groupby('Branch_Code')['Balance'].mean().sort_values(ascending=False).head(5)

# Calculate the highest average transaction counts by branch
highest_avg_transaction_count = df.groupby('Branch_Code')['Transaction_Count'].mean().sort_values(ascending=False).head(5)

# Display the results
print("Branches with the highest average balances:")
print(highest_avg_balance)

print("\nBranches with the highest average transaction counts:")
print(highest_avg_transaction_count)
```

### 6. What factors (e.g., low balance, fewer transactions) are associated with dormant accounts?

```python
# Separate the data into dormant and non-dormant accounts
dormant_accounts = df[df['Status'] == 'Dormant']
non_dormant_accounts = df[df['Status'] != 'Dormant']

# Compare average balance and transaction count
dormant_stats = {
    "Average_Balance": dormant_accounts['Balance'].mean(),
    "Average_Transactions": dormant_accounts['Transaction_Count'].mean()
}

non_dormant_stats = {
    "Average_Balance": non_dormant_accounts['Balance'].mean(),
    "Average_Transactions": non_dormant_accounts['Transaction_Count'].mean()
}

# Create a comparison DataFrame
comparison_df = pd.DataFrame([dormant_stats, non_dormant_stats], index=['Dormant', 'Non-Dormant'])

# Display the comparison
print("Factors Associated with Dormant Accounts:")
print(comparison_df)
```

### 7. Is there a correlation between the time of the last transaction and account activity levels ?

```python
# Extract the hour from the Last_Transaction_Time
df['Last_Transaction_Hour'] = pd.to_datetime(df['Last_Transaction_Time'], format='%H:%M:%S').dt.hour

# Calculate the correlation between Last_Transaction_Hour and Transaction_Count
correlation = df['Last_Transaction_Hour'].corr(df['Transaction_Count'])

# Display the correlation result
print(f"Correlation between the time of the last transaction (hour) and account activity levels: {correlation}")
```

### 9. Do younger or older account holders show higher transaction activity?

```python
from datetime import datetime

# Ensure the Age column is calculated
current_year = datetime.now().year
df['Age'] = current_year - pd.to_datetime(df['Date_of_Birth']).dt.year

# Divide account holders into age groups
age_bins = [18, 30, 40, 50, 60, 70, 80, 90]
age_labels = ['18-30', '31-40', '41-50', '51-60', '61-70', '71-80', '81-90']
df['Age_Group'] = pd.cut(df['Age'], bins=age_bins, labels=age_labels)

# Calculate the average transaction count for each age group
age_group_activity = df.groupby('Age_Group')['Transaction_Count'].mean()

# Display the results
print("Average Transaction Count by Age Group:")
print(age_group_activity)

# Visualization
plt.figure(figsize=(10, 6))
sns.barplot(x=age_group_activity.index, y=age_group_activity.values, palette="viridis")
plt.title('Average Transaction Activity by Age Group', fontsize=16)
plt.xlabel('Age Group', fontsize=12)
plt.ylabel('Average Transaction Count', fontsize=12)
plt.tight_layout()

# Show the plot
plt.show()
```

### 10. How have account openings varied over time? Are there seasonal or yearly trends ?

```python
# Convert the Account_Creation_Date to datetime format
df['Account_Creation_Date'] = pd.to_datetime(df['Account_Creation_Date'])

# Extract year and month for analysis
df['Year'] = df['Account_Creation_Date'].dt.year
df['Month'] = df['Account_Creation_Date'].dt.month

# Group by year and count the number of accounts opened
yearly_trends = df.groupby('Year').size()

# Group by month (across all years) to find seasonal trends
monthly_trends = df.groupby('Month').size()

# Display trends
print("Yearly Account Opening Trends:")
print(yearly_trends)

print("\nSeasonal Trends in Account Openings (Monthly):")
print(monthly_trends)
```

### 11. What characteristics are common among accounts with the highest balances?

```python
# Define the threshold for highest balances (e.g., top 10%)
threshold = df['Balance'].quantile(0.9)

# Filter for accounts with balances in the top 10%
high_balance_accounts = df[df['Balance'] >= threshold]

# Summarize characteristics of high-balance accounts
summary_stats = high_balance_accounts.describe(include='all')

# Group by account type to understand distribution
account_type_distribution = high_balance_accounts['Account_Type'].value_counts(normalize=True) * 100

# Group by currency to understand distribution
currency_distribution = high_balance_accounts['Currency'].value_counts(normalize=True) * 100

# Display results
print("Summary Statistics for Accounts with the Highest Balances:")
print(summary_stats)

print("\nAccount Type Distribution Among High-Balance Accounts:")
print(account_type_distribution)

print("\nCurrency Distribution Among High-Balance Accounts:")
print(currency_distribution)
```

### 12. Are certain email domains (e.g., Gmail, Yahoo) more prevalent among account holders?

### 13. Are there geographic trends based on phone number prefixes?

```python
# Extract phone number prefixes (assuming the first 3 digits are the prefix)
df['Phone_Prefix'] = df['Phone_Number'].astype(str).str[:3]

# Count the occurrences of each prefix
prefix_distribution = df['Phone_Prefix'].value_counts()

# Display the most common phone prefixes
print("Phone Number Prefix Distribution:")
print(prefix_distribution)

# Analyze prefix trends (if location data is available, correlate prefix with location)
if 'Location' in df.columns:
    prefix_location_trends = df.groupby(['Phone_Prefix', 'Location']).size().reset_index(name='Count')
    print("\nGeographic Trends by Phone Number Prefix:")
    print(prefix_location_trends)
```

### 15. Do account holders with specific currencies tend to have higher average transaction counts?

```python
# Check if required columns exist
if 'Currency' in df.columns and 'Transaction_Count' in df.columns:
    # Group by Currency and calculate average transaction count
    avg_transaction_counts = df.groupby('Currency').agg(
        Avg_Transaction_Count=('Transaction_Count', 'mean'),
        Total_Accounts=('Transaction_Count', 'count')
    ).sort_values(by='Avg_Transaction_Count', ascending=False)

    # Display the results
    print("Average Transaction Counts by Currency:")
    print(avg_transaction_counts)
else:
    print("One or more required columns ('Currency', 'Transaction_Count') are missing in the dataset.")
```

## PLOTTING OF PIVOT TABLE

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Create a pivot table summarizing balances and transaction counts by Account_Status and Currency
pivot_table = pd.pivot_table(
    df,
    values='Balance',  # Use Balance as the numerical value to visualize
    index='Status',  # Rows
    columns='Currency',      # Columns
    aggfunc='mean',          # Aggregation function
    margins=False            # Exclude totals for better heatmap readability
)

# Display the pivot table
print("Pivot Table (Average Balance by Account Status and Currency):")
print(pivot_table)

# Visualize the pivot table using a heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(
    pivot_table,
    annot=True,              # Annotate each cell with the value
    fmt=".2f",               # Format values to 2 decimal places
    cmap="YlGnBu",           # Color map
    cbar_kws={'label': 'Average Balance'},  # Color bar label
    linewidths=0.5           # Line width between cells
)

# Customize the heatmap
plt.title('Heatmap of Average Balance by Account Status and Currency', fontsize=16)
plt.xlabel('Currency', fontsize=12)
plt.ylabel('Account Status', fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()

# Show the heatmap
plt.show()
```
